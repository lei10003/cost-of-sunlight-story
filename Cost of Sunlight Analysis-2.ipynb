```python
ids = []
zipcode = []
unit_number = []
building_class =[]
beds = []
size = []
rents = []
light_cat = []
lights = []

for dictionary in dictionaries:
    for id in dictionary['data']['searchResult']['poi']:
        for insight in id['insights']['insights']:
            if insight['category'] == 'naturalLight':
                ids.append(id['id'])
                zipcode.append(id['addressDetails']['zipcode'])
                unit_number.append(id['addressDetails']['unitNumber'])
                building_class.append(id['buildingClass'])
                beds.append(id['beds'])
                size.append(id['area'])
                rents.append(id['price'])
                light_cat.append(insight['tradeoff']['value'])
                lights.append(insight['tradeoff']['value'] + f" and the rent is {id['price']}")
        if len(id['insights']['insights']) == 0 or all(insight['category'] != 'naturalLight' for insight in id['insights']['insights']):
            ids.append(id['id'])
            zipcode.append(id['addressDetails']['zipcode'])
            unit_number.append(id['addressDetails']['unitNumber'])
            building_class.append(id['buildingClass'])
            beds.append(id['beds'])
            size.append(id['area'])
            rents.append(id['price'])
            light_cat.append(f"No sunlight")
            lights.append(f"No sunlight data available and the rent is {id['price']}")
```

```python
data = {
    'listing_path': ids,
    'location': zipcode,
    'unit_number': unit_number,
    'building_class': building_class,
    'beds': beds,
    'size':size,
    'rent': rents,
    'sun_exposure_cat': light_cat,
    'sun_exposure': lights
}

df_sun_exposure = pd.DataFrame(data)
```

```python
# Define the new column names mapping
column_names = {
    'lives alone.1': 'lives alone 18-34',
    'householder living with spouse or spouse of householder.1': 'householder with spouse 18-34',
    'householder living with unmarried partner or unmarried partner of householder.1': 'householder with partner 18-34',
    'child of householder.1': 'child of householder 18-34',
    'other relatives.1': 'other relatives 18-34',
    'other nonrelatives.1': 'other nonrelatives 18-34',
    'lives alone.2': 'lives alone 35-64',
    'householder living with spouse or spouse of householder.2': 'householder with spouse 35-64',
    'householder living with unmarried partner or unmarried partner of householder.2': 'householder with partner 35-64',
    'child of householder.2': 'child of householder 35-64',
    'other relatives.2': 'other relatives 35-64',
    'other nonrelatives.2': 'other nonrelatives 35-64',
    'lives alone.3': 'lives alone 65+',
    'householder living with spouse or spouse of householder.3': 'householder with spouse 65+',
    'householder living with unmarried partner or unmarried partner of householder.3': 'householder with partner 65+',
    'child of householder.3': 'child of householder 65+',
    'other relatives.3': 'other relatives 65+',
    'other nonrelatives.3': 'other nonrelatives 65+'
}

# Rename the columns
df_acs_selected_zipcodes.rename(columns=column_names, inplace=True)
```

```python
df_acs_selected_zipcodes
```

```python
# Define age groups for the columns
age_groups = ['18-34', '35-64', '65+']

# Initialize an empty list to store each reshaped DataFrame
dfs = []

# Loop over each age group
for age in age_groups:
    # Select the columns related to the current age group
    cols = [col for col in df_acs_selected_zipcodes.columns if age in col]
    
    # Reshape the DataFrame
    df_melted = df_acs_selected_zipcodes.melt(id_vars='zipcode', value_vars=cols, var_name='living_arrangement', value_name='count')
    
    # Add the age group column
    df_melted['age_group'] = age
    
    # Append the reshaped DataFrame to the list
    dfs.append(df_melted)

# Concatenate all reshaped DataFrames
df_tidy_acs = pd.concat(dfs, ignore_index=True)

# Extract the living arrangement from the 'living_arrangement' column
df_tidy_acs['living_arrangement'],df_tidy_acs['age_group'] = zip(*df_tidy_acs['living_arrangement'].str.rsplit(n=1))
```
